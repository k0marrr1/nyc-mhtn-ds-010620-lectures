```python
import csv
```

```python
with open ('data.csv') as f:
    reader = csv.DictReader(f)
    
    my_list = [ ]
    for item in reader:
        my_list.append(item)
        print (item)
```

```python
RS_dict = []
for item in my_list:
    RS_dict.append(dict(item))
    print(RS_dict)
```

```python
RS_dict
```

```python
def find_by_name(album):
    for item in RS_dict:
        if item['album'] == album:
            return item
        else:
            return False
```

```python
find_by_name('Omar')
```

```python
def find_by_rank(rank):
    for item in RS_dict:
        if item ['number'] == rank:
            return item
        else:
            if rank > '500':
                return False
```

```python
find_by_rank('500')
```

```python
def find_by_year(year):
    album_year = []
    for item in RS_dict:
        if item['year'] == year:
            album_year.append(item['album'])
    return album_year
```

```python
find_by_year('1971')
```

```python
def find_by_years(year1, year2):
    album_list = []
    for item in RS_dict:
        if item['year'] >= year1 and item['year'] <= year2:
            album_list.append(item['album'])
    return album_list
        
```

```python
find_by_years('1960','1965')
```

```python
def find_by_ranks(rank1, rank2):
    album_list = []
    for item in RS_dict:
        if int(item['number']) >= rank1 and int(item['number']) <= rank2:
            album_list.append(item['album'])
    return album_list
```

```python
find_by_ranks(1,2)
```

```python
def all_titles(album):
    album_title_list = []
    for item in RS_dict:
        if item['album'] == album:
            album_title_list.append(item['album'])
    return album_title_list
```

```python
all_titles('Hotel California')
```

```python
def all_artists():
    artist_list = []
    for item in RS_dict:
        artist_list.append(item['artist'])
    return artist_list
```

```python
all_artists()
```

```python
# Artists with the most albums - Returns the artist with the highest amount of albums on the list of top albums
list#
from collections import Counter, defaultdict

artist_list = all_artists()

d = defaultdict(int)
for artist in artist_list:
    d[artist] += 1
d
```

```python
artist_dict = dict(d)
artist_dict
```

```python
def artist_most_albums():
    a_list = []
    list_of_vals = list(artist_dict.values())
    list_of_keys = list(artist_dict.keys())
    for idx, val in enumerate(list_of_vals):
        if val == max(list_of_vals):
            a_list.append(list_of_keys[idx])
    return a_list
artist_most_albums()
```

```python
def pop_word():
    word_list = []
    top_words = []
    from collections import Counter
    for item in RS_dict:
        word_list.extend(item['album'].split(" "))
    word_dict = Counter(word_list)
    vals = list(word_dict.values())
    keys = list(word_dict.keys())
    for idx, val in enumerate(vals):
        if val == max(vals):
            top_words.append(keys[idx])
    return top_words
pop_word()
```
